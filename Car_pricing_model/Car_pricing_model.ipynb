## Обучение моделей

Функции разделения (и кодировки) данных; тренировки моделей; подбора гиперпараметров:

```python
def split_data(dataframe, scaled=True, poly=False, ohe=False, ordinal=False,
               val_test_size=0.4, random_state=12, poly_degree=2):
    '''
    Фукнция делит датафрейм на 3 выборки: тренировочную, валидационную и тестовую;
    а также опционально применяет One-Hot и Ordinal кодирование данных
    '''
    for c in cat_cols:  
        dataframe[c] = dataframe[c].astype('category')
        
    features = dataframe.drop(target_col, axis=1)
    target = dataframe[target_col]
    
    features_train, features_valtest, target_train, target_valtest = train_test_split(
        features, target, test_size=val_test_size, random_state=random_state)
    
    features_val, features_test, target_val, target_test = train_test_split(
        features_valtest, target_valtest, test_size=0.5, random_state=random_state)
    
    fts = [features_train, features_val, features_test]
    tgs = [target_train.values.ravel(), target_val.values.ravel(), target_test.values.ravel()]
        
        
    if scaled:       # нормализация
        scaler = StandardScaler()      
        scaler.fit(fts[0][num_cols])
        for f in range(len(fts)):          
            scaled = pd.DataFrame(scaler.transform(fts[f][num_cols]), columns=fts[f][num_cols].columns, index=fts[f].index)
            fts[f] = pd.concat([fts[f].drop(num_cols, axis=1), scaled], axis=1)
            for n in num_cols:
                fts[f][n] = pd.to_numeric(fts[f][n], downcast='float')
        
    if poly:         # генерирование новых признаков из-за нелинейности некоторых парных диаграмм распределения
        encoder = PolynomialFeatures(degree=poly_degree)
        encoder.fit(fts[0][num_cols])
        for f in range(len(fts)):
            encoded_poly = pd.DataFrame(encoder.transform(fts[f][num_cols]),
                                  columns=encoder.get_feature_names_out(), index=fts[f].index)
            fts[f] = pd.concat([fts[f].drop(num_cols, axis=1), encoded_poly], axis=1)
            
    if ohe:          # One-Hot кодирование
        encoder = OneHotEncoder(drop='first', handle_unknown='ignore', dtype='int8')
        encoder.fit(fts[0][cat_cols])
        for f in range(len(fts)):
            encoded_ohe = pd.DataFrame(encoder.transform(fts[f][cat_cols]).toarray(),
                                         columns=encoder.get_feature_names_out(), index=fts[f].index)
            fts[f] = pd.concat([fts[f].drop(cat_cols, axis=1), encoded_ohe], axis=1)
            
    if ordinal:       # Порядковое кодирование
        encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1, dtype='int16')
        encoder.fit(fts[0][cat_cols])
        for f in range(len(fts)):
            encoded_ord = pd.DataFrame(encoder.transform(fts[f][cat_cols]),
                                         columns=fts[f][cat_cols].columns, index=fts[f].index)
            fts[f] = pd.concat([fts[f].drop(cat_cols, axis=1), encoded_ord], axis=1)
          
    return *fts, *tgs
```

```python
def train_model(model_name, params, features_train, features_test, target_train, target_test, **kwargs):
    '''Тренирует модель (fit), и предсказывает (predict)
    Возвращает натренированную модель, показатель качества предсказаний RMSE
    и время, затраченные на тренировку и предсказание
    '''
    model = model_name(**params, **kwargs)
    
    start_train = time()
    model.fit(features_train, target_train)
    stop_train = time()
    
    start_infer = time()
    pred_test = model.predict(features_test)
    stop_infer = time()
    
    rmse = mean_squared_error(pred_test, target_test) ** 0.5
    
    return model, rmse, stop_train - start_train, stop_infer - start_infer

def finetune(model_name, parameters, features, target, verbose=3, jobs=-1, **kwargs):
    '''Фукнция подбора гиперпараметров
    На вход принимает класс модели, решетку гиперпараметров для поиска, признаки и целевой признак
    Возвращает лучшую из натренерованных моделей, её гиперпараметры и её луший результат по метрике RMSE
    '''
    cv = RandomizedSearchCV(model_name(**kwargs), param_distributions=parameters,
                      scoring='neg_mean_squared_error', cv=3, refit=True, verbose=verbose, n_jobs=jobs)
    cv.fit(features, target)  
    
    return cv.best_estimator_, cv.best_params_, np.sqrt(-cv.best_score_)

def print_cv_results(params, rmse):
    '''Выводит результаты кросс-валидации на экран'''
    print()
    print(f'Лучшие параметры модели: {params}')
    print(f'(train, CV) лучшее RMSE: {rmse:.2f}')
    print()
```

### Формирование рейтинга моделей

Все полученные значения моделей находятся в обратных метриках (чем меньше, тем лучше), чтобы перейти к рейтингу, нужно перейти к прямым метрикам, для этого домножаю на -1.

Также, чтобы значения были сопоставимы, нормализую все колонки:


```python
normalized_stats = (final_stats * -1).apply(lambda x: (x - x.mean()) / x.std())
normalized_stats
```

Привожу значения к рейтингу от 0 до 10 через сигмоиду:

```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))

ratings = normalized_stats.apply(sigmoid) * 10
ratings
```

Допустим, клиент определил следующий приоритет характеристик у моделей: важнее всего скорость предсказаний, затем точность предсказаний, затем время обучения. Для примера, возьму такие веса: скорость - 0,5; качество - 0,3; время - 0,2:

```python
weights = pd.DataFrame({'': 'веса',
                        '(val) RMSE': [0.3],
                        'Training time': [0.2],
                        'Inference time':[0.5]}).set_index('')
weights
```

Перемножением матрицы рейтингов на веса можно получить итоговый рейтинг модели:

```python
model_rating = np.dot(ratings.values, weights.values.T)
final_rating = pd.DataFrame(model_rating.ravel().round(2), columns=['Rating'], index=final_stats.index).sort_values(by='Rating', ascending=False)
final_rating
```
